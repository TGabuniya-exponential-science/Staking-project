```python
####Load the libraries####
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
####Load the data####
mI = pd.read_excel("maximumsIa.xlsx")
mII = pd.read_excel("maximumsIIa.xlsx")
mIII = pd.read_excel("maximumsIIIa.xlsx")
mIV= pd.read_excel("maximumsIVa.xlsx")
```

```python
plt.plot(mI['r'])
plt.plot(mII['r'])
plt.plot(mIII['r'])
plt.plot(mIV['r'])
plt.legend(['Scenario I','Scenario II','Scenario III','Scenario IV',])
plt.xticks(ticks = [0, 1, 2, 3, 4], labels = [-1, -0.5, 0, 0.5, 1])
plt.xlabel(r'$\rho$')
plt.ylabel('r (decimals)')
plt.tight_layout()
plt.savefig('rOpt.jpg')
plt.show()
```

```python
plt.plot(mI['t'])
plt.plot(mII['t'])
plt.plot(mIII['t'])
plt.plot(mIV['t'])
plt.xticks(ticks = [0, 1, 2, 3, 4], labels = [-1, -0.5, 0, 0.5, 1])
plt.xlabel(r'$\rho$')
plt.ylabel('t (years)')
plt.legend(['Scenario I','Scenario II','Scenario III','Scenario IV',])
plt.tight_layout()
plt.savefig('tOpt.jpg')
plt.show()
```
